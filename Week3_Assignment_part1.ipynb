### 1. Import necesary libraries

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

### 2. Get the HTML for the website

```python
request = requests.get('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M').text
```

### 3. Scrape the website using BeautifulSoup

```python
soup = BeautifulSoup(request, 'lxml')
print(soup.prettify())
```

### 4. Since our soup is cooked it's time fetch the table

```python
table = soup.find('table')
print(table.prettify())
```

### 5.  Now that we found the table from our HTML we can now print the data from the table

```python
info = []
for row in table.find_all('tr'):
    for cell in row.find_all('td'):
        info.append(cell.text)
```

```python
info
```

### 6. We shall pick the data one by one from the table in HTML and insert it into the dataframe

```python
postcode = []
borough = []
Neighborhood = [] #lists defined
for i in range(0,len(info),3):
    if info[i+1] == 'Not assigned' :   
        pass #ignoring all the cells that have Not assigned in the Borough field
    else :
        postcode.append(info[i]) #creating the postcode list
        borough.append(info[i+1]) #creating the borough list
        if info[i+2] == 'Not assigned\n' : 
            Neighborhood.append(info[i+1]) #if Neighborhood is Not assigned then make it same as the borough
        else :
            Neighborhood.append(info[i+2].strip('\n')) #otherwise keep it as it is
```

```python
print(postcode)
print("******************************************************************************************************************")
print(borough)
print("******************************************************************************************************************")
print(Neighborhood)
```

#### all our data looks good

### 7. Defining the dataframe

```python
# define the dataframe columns
column_names = ['PostalCode','Borough','Neighborhood']

# instantiate the dataframe
neighborhoods = pd.DataFrame(columns=column_names)
neighborhoods
```

### Checking the lengths of the lists so far(whether they are of same length)

```python
print(len(postcode))
print(len(borough))
print(len(Neighborhood))
```

```python
for i in range(0,len(postcode)-1):
    j = i+1
    while postcode[i]==postcode[j]:
        Neighborhood[i] = Neighborhood[i] + ',' + Neighborhood[j] 
        j+=1    
        
```

```python
def reverseWords(input): 
    inputWords = input.split(",")
    inputWords=inputWords[-1::-1]
    output = ','.join(inputWords)
    return output
```

```python
reverseWords('Harbourfront,Regent Park')
```

```python
neighborhood = []
for i in Neighborhood:
    neighborhood.append(reverseWords(i))
```

### Let's Check the Neighborhood list now

```python
for i in neighborhood:
    print(i)
```

```python
print(len(postcode))
print(len(borough))
print(len(neighborhood))
```

### 8. Insert the lists into a dataframe 

```python
for a,b,c in zip(postcode,borough,neighborhood):    
    neighborhoods = neighborhoods.append({'PostalCode' : a,
                                          'Borough': b,
                                          'Neighborhood': c}, ignore_index=True)
```

```python
neighborhoods
```

### 9. Now if we filter out the duplicate columns and keep the first occurrences we can get the dataframe as asked in the question

```python
neighborhoods.drop_duplicates(subset='PostalCode', keep='first', inplace=True)
```

```python
neighborhoods = neighborhoods.reset_index()
```

```python
neighborhoods = neighborhoods.drop('index',axis = 1)
```
