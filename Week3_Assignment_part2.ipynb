# 4. Making lists of Latitides and Longitudes from the given dataset before inserting into our actual dataset called "data".

```python
lat = [] 
lon = [] #initialize the lists
for i in range(len(data['PostalCode'])): #iterate through the postal codes of our dataset
    for j in range(len(postcode['Postal Code'])): #iterate through the postal codes of the given dataset 
        if data.loc[i,'PostalCode'] == postcode.loc[j,'Postal Code']: #checking for matching postal codes
            lat.append(postcode.loc[j,'Latitude']) #appending to the lat list for matches 
            lon.append(postcode.loc[j,'Longitude']) #appending to the lon list for matches
```

# 5. Insert the lists into the dataframe as columns

```python
data['Latitude'] = lat
data['Longitude'] = lon
```

```python
data = data.drop('Unnamed: 0',axis = 1)
```
